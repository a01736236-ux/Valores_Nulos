```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .xlsx sin indice
data = pd.read_excel('Gastos y costos 20-23.xlsx',skiprows=5)
data
```

```python
#Comprobamos valores nulos 
data.isnull().sum()
```

```python
#Realizamos una copia del dataframe
data1=data.copy()
```

```python
#Poliza se registra como "no hay dato" ya que no se puede suplir por una inventada
data1["POLIZA"] =data1["POLIZA"].fillna("no hay dato") 
```

```python
#Comprobamos
data1.isnull().sum()
```

```python
data1["FOLIO"] =data1["FOLIO"].fillna(method="ffill") 
```

```python
data1["GASTO"] =data1["GASTO"].fillna("GASTO")
```

```python
data1["TC"] =data1["TC"].fillna(method="ffill") 
```

```python
data1["IMPORTE"] =data1["IMPORTE"].fillna(method="ffill") 
```

```python
data1["IVA"] =data1["IVA"].fillna(method="ffill") 
```

```python
data1["TIPO"] =data1["TIPO"].fillna(method="ffill") 
```

```python
#Comprobamos
data1.isnull().sum()
```
